```python
#Create a program where the user can play a game of rock, paper, scissors against a computer. 
#Set a play choice (rock, paper, or scissors) for the computer player and during the program, 
#request for the user to input their play choice.

#The rules in Rock, Paper, Scissors are:

#Rock beats scissors
#Scissors beats paper
#Paper beats rock
#If the user wins, then display "You win!" 
#If they lose against the computer, then display "You lose." 
#If the computer and the user have the same play choice, display "It's a draw."

#BONUS: Randomize the play choice for the computer (hint: Python has a library called random). 
#Also ask the user if they would like to play again. 
#If they do, start the game again but if not, thank the user for playing the game.

#Date 10-31-2023


import random as r

#print(r.randint(1,3))

#print (ComputerChoiceNum)

def SetComputerChoice() :
    ComputerChoiceNum = r.randint(1,3)
    
    if ComputerChoiceNum == 1 :
        ComputerChoice = "scissors"
    elif ComputerChoiceNum == 2 :
        ComputerChoice = "paper"
    elif ComputerChoiceNum == 3 :
        ComputerChoice = "rock"
                                  
    return (ComputerChoice)

def GetUserChoice() :
    
    print ("Please Input you choice of Rock/Paper/Scissors")
    UserChoice = input()
    lUserChoice = UserChoice.lower()

    return (lUserChoice)

def GetDone () :
    print ("Please enter 0 to stop playing else enter any other number ONLY")
    GetDoneChoice = int(input())
    
    return(GetDoneChoice)
    

#initalize done
done = -1

# The game continues until a 0 is input

while (done != 0) :
    
    ComputerHas = SetComputerChoice()
    UserHas = GetUserChoice()
    
    print("Me, the computer choose -", ComputerHas)
    print ("You choose -:", UserHas)
    
    if (ComputerHas == UserHas) :
        print ("I, the computer and you have the same choice - ", ComputerHas , "- It's a draw.\n")
    

    #1. User has Rock
    elif (UserHas == "rock" and ComputerHas == "paper") :
        print ("I, the computer chose -", ComputerHas, "and you have - ", UserHas , "- You Lose.\n")
    elif (UserHas == "rock" and ComputerHas == "scissors") :
        print("I, the computer chose -", ComputerHas, " and you have - ", UserHas , "- You Win!\n")

    #2. User has Paper   
    elif (UserHas == "paper" and ComputerHas == "scissors") :
        print ("I, the computer chose -", ComputerHas, "and you have - ", UserHas , "- You Lose.\n")
    elif (UserHas == "paper" and ComputerHas == "rock") :
        print("I, the computer chose -", ComputerHas, " and you have - ", UserHas , "- You Win!\n")
    
    #2. User has scissors   
    elif (UserHas == "scissors" and ComputerHas == "rock") :
        print ("I, the computer chose -", ComputerHas, "and you have - ", UserHas , "- You Lose.\n")
    elif (UserHas == "scissors" and ComputerHas == "paper") :
        print("I, the computer chose -", ComputerHas, " and you have - ", UserHas , "- You Win!\n")
        
    done = GetDone()

print ("Goodbye, Nice playing with you")
```
