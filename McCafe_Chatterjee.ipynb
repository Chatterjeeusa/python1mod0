```python
#Your instructors love coffee so much that they decided to open a cafe! Of course, we will need the help of our 
#students to develop the ordering app that the customers can use.

#On the cafe menu, there are three categories for the items we sell - coffee, tea, and pastries.

#Coffee
#    Espresso - $1.50
#    Americano - $2.00
#    Caffe latte - $2.25
#    Mocha - $2.25
#    Cappucino - $2.50
#    Macchiato - $2.50
#Tea
#    White tea - $1.50
#    Green tea - $2.00
#    Oolong tea - $2.25
#    Black tea - $2.25
#    Earl Grey - $2.50
#    Matcha - $2.50
#Pastries
#    Croissant - $5.50
#    Cinnamon roll - $4.50
#    Banana bread - $4.25
#    Pumpkin bread - $3.75
#    Sticky bun - $4.25
#   Apple turnover - $3.75

#When your program first starts, you must greet the customers by saying, "Welcome to MC Cafe!" 
#Then ask the customer if they would like to see the menu. If yes, then display the menu items and prices.

#Then ask the user if they would like to order. If yes, then prompt the customer to enter an item and 
#the quantity that they want. After they input an item, then ask them if they would like to order another item.

#Once your customers are finished with the order input, then display an order receipt that lists the 
#items that they ordered and the quantity. Then at the bottom of the receipt, display the subtotal 
#(the amount of the item before tax), the amount in sales tax (our business is in Maryland, so state tax is 6%), 
#and finally their grand total (subtotal plus sales tax).

#At the end of their transaction, display a message to our customers saying, "Thank you for visiting MC Cafe!"

#BONUS: Add a feature that tells a customer when an item they attempted to order is not on the menu.

import time
 


Menu = {}

Menu = {"Coffee" : {
        "Espresso": 1.50, 
        "Americano" : 2.00, 
        "Caffe latte" :2.25, 
        "Mocha" : 2.25, 
        "Cappucino" :2.50, 
        "Macchiato" : 2.50
        },
        
    "Tea": { 
         "White tea" : 1.50,
        "Green tea" : 2.00,
        "Oolong tea" : 2.25,
        "Black tea" : 2.25,
        "Earl Grey" : 2.50,
        "Matcha" : 2.50
        },

    "Pastries" : {
        "Croissant" : 5.50,
        "Cinnamon roll" : 4.50,
        "Banana bread" : 4.25,
        "Pumpkin bread" : 3.75,
        "Sticky bun" : 4.25,
        "Apple turnover" : 3.75
        }
    }

#borrowed code from Internet
def _lowercase(obj):
    """ Make dictionary lowercase """
    if isinstance(obj, dict):
        return {k.lower():_lowercase(v) for k, v in obj.items()}
    elif isinstance(obj, (list, set, tuple)):
        t = type(obj)
        return t(_lowercase(o) for o in obj)
    elif isinstance(obj, str):
        return obj.lower()
    else:
        return obj
    

def TakeOrder(lmenu) :
    #print ("Taking Order")
    #Ask Customer to order an item
    comparingMenu = {}
    comparingMenu = _lowercase(lmenu)
    
    print("Iteams available are:")
    
    #print the main menu items for customer to pick ---- Coffee, Tea, Pastries
    for key in comparingMenu.keys() :   # assigns key and value variable simultaneously
        print (f"{key}\t")
    orderedItem = input("Please enter one item of the above mentioned menu item for it's detail menu\t")

    orderedItem = orderedItem.lower()
    print (f"Ordering {orderedItem} \n")
 
    OrderedItemsList = {}
    itemFoundFlag = False
    
    #Traverse the outer keys of the dictionary - Coffee, tea and pastries
    for key, value in comparingMenu.items() :   # assigns key and value variable simultaneously
        if orderedItem == key :
            #print ("Item found in")
            #print(comparingMenu.keys())
           
            #Traverse the inner keys of the dictionary - details of coffee and others...
            for key1, value1 in comparingMenu[key].items() :
                print(f"\t{key1}\t: {value1}")
            
            specificItem = input(f"Please enter what kind of above mentioned {orderedItem} would you like\t")
            
            for key1, value1 in comparingMenu[key].items() :
                if specificItem == key1 : # check if item provided with the customer exits in the dictionary
                    checkForNum = 0
                    
                    while checkForNum == 0 : #check how many units customer wants
                        HowMany = (input(f"Please enter ONLY NUMBERS how many {specificItem} would you like to order\t"))
                        
                        if HowMany.isdigit() == True : #check to see if a number is provided
                            checkForNum = 1
                            HowMany = float(HowMany)
                        
                    #print(f"\t{key1}\t: is going to cost {value1} each") #let the customer know how the cost for items 
                    OrderedItemsList[key1] =  HowMany * value1
                    
                    itemFoundFlag = True
            if itemFoundFlag == False : # item entered does not exists in the detailed dictionary
                    print(f"No such {specificItem} for {orderedItem}s")
        
        
    if itemFoundFlag == False : #entered item does not exists in the "keys" of dictionary
        print(f"No such ({orderedItem}) on the menu")
    #else :
        #print (f"ordered items are {OrderedItemsList.items()}\n\n")
            
    return (OrderedItemsList)


    
def FormattedMenu(lmenu) :
    
    #Another way of printing Menu
    for key, value in lmenu.items() :                     # assigns key and value variable simultaneously
        print(f"{key}") #\tvalue: {value}"
        for key1, value1 in lmenu[key].items() :
            print(f"\t{key1}\t: {value1}")

  
    
def DisplayReceipt(llist) :
    
    SubTotal = 0
    TaxRate = .06
    
    curr_time = time.strftime("Date:%D  Time:%H:%M:%S", time.localtime())
   
 
    
    print("\n******** Receipt ******** ")
    print (f"{curr_time}\n")
    print("------------------------------")
    for key, value in llist.items() :
            print(f"{key}:\t\t\t${value:.2f}")
            SubTotal += value
    print (f"\nSubTotal = \t\t${SubTotal:.2f}")
    Tax = SubTotal * TaxRate
    print (f"MD Sales Tax = \t\t${Tax:.2f}")
    print("------------------------------")
    Total = SubTotal+Tax
    print (f"Total = \t\t${Total:.2f}")
    print("------------------------------")
    
    
###### START

print ("Welcome to MC Cafe!")

OrderTaken = False
OrderList = {}

CustomerResponse = input("Please ONLY enter Y or N - Would like to see the menu? y/n\n")
CustomerResponse = CustomerResponse.lower()
   
while CustomerResponse != "y" and CustomerResponse != "n" :

    print("Please ONLY enter Y or N")
    CustomerResponse = input("Please ONLY enter Y or N - Would like to see the menu? y/n\n")
    CustomerResponse = CustomerResponse.lower()
            
if CustomerResponse.lower() == "y" :
    print(FormattedMenu(Menu))


#Ask to take the order
CustomerResponse2 = input("Please ONLY enter Y or N - Would like to order? y/n\n")
CustomerResponse2 = CustomerResponse2.lower()

while CustomerResponse2 != "y" and CustomerResponse2 != "n" :

    print("Please ONLY enter Y or N")
    print(f"The customer response is - {CustomerResponse2}")
    
    CustomerResponse2 = input("Please ONLY enter Y or N - Would like to see the menu? y/n\n")
    CustomerResponse2 = CustomerResponse2.lower()


if CustomerResponse2.lower() == "y" :
    OrderList = TakeOrder(Menu)
    OrderTaken = True
    
    flag = "y"
    while flag == "y" :
        CustomerResponse3 = input("Would like to order another item? y/n\n")
        CustomerResponse3 = CustomerResponse3.lower()
        
        while CustomerResponse3 != "y" and CustomerResponse3 != "n" :

            print("Please ONLY enter Y or N")
            CustomerResponse3 = input("Please ONLY enter Y or N - Would like to see the menu? y/n\n")
            CustomerResponse3 = CustomerResponse3.lower()
                    
        
        if CustomerResponse3.lower() == "y" :
            OrderList.update(TakeOrder(Menu))
            OrderTaken = True
        else :
            flag = "N"

#only print receipt if any item was ordered
if len(OrderList) != 0 :
    DisplayReceipt(OrderList)


print("\nThank you for visiting MC Cafe!")
```
